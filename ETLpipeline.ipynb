```python
# 1nf normalization split the customername to firstname and lastname 

import pandas as pd

# Function to split names into first and last names
def split_name(name):
    parts = name.split()
    if len(parts) >= 2:
        first_name = parts[0]
        last_name = parts[-1]
    elif len(parts) == 1:
        first_name = parts[0]
        last_name = ''
    else:
        first_name = ''
        last_name = ''
    return pd.Series([first_name, last_name])

# Apply the function to the 'Name' column and create new columns
df[['First Name', 'Last Name']] = df['CustomerName'].apply(split_name)

print(df)
```

```python
df.columns
```

```python
#customer table 
customer = df[['CustomerID','First Name','Last Name','CustomerEmail']].drop_duplicates()
customer
```

```python
# prdouct tables
product = df[['ProductID', 'ProductName','PaymentMethod']].drop_duplicates()
product
```

```python
# category table
category = df[[ 'CategoryID', 'CategoryName']].drop_duplicates()
category.head(3)
```

```python
#order table
order = df[['OrderID','OrderDate', 'ShippingDate', 'DeliveryDate','ShippingAddress', 'OrderStatus']].drop_duplicates()
order
```

```python
facttable = df.merge(customer, on=['CustomerID', 'First Name', 'Last Name', 'CustomerEmail'], how='left') \
              .merge(product, on=['ProductID', 'ProductName', 'PaymentMethod'], how='left') \
              .merge(order, on=['OrderID','OrderDate', 'ShippingDate', 'DeliveryDate', 'ShippingAddress', 'OrderStatus'], how='left') \
              .merge(category, on=['CategoryID', 'CategoryName'], how='left')
facttable['factid'] = range(1,len(facttable)+1)
facttable=facttable[['factid','CustomerID','ProductID','OrderID','CategoryID','Quantity', 'UnitPrice', 'TotalPrice','Rating', 'Review'\
                     ]].drop_duplicates()
facttable
```

```python
#save table in csv file
customer.to_csv('customer',index=False)
product.to_csv('product',index=False)
order.to_csv('order',index=False)
category.to_csv('category',index=False)
facttable.to_csv('facttable',index=False)
```
